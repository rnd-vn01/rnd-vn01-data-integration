```python
import numpy as np
import pandas as pd
```

```python
sheets = ["Nhân (V)", "Tân (V)", "Chương (V)", "Trang (V)"]
```

```python
# Read data
df_data = pd.read_excel("DATA COLLECTION (60 IMPORTANT POINTS).xlsx", sheet_name=sheets)
for sheet in df_data:
    df_data[sheet] = df_data[sheet].iloc[:, :7]
```

```python
df_data["Tân (V)"].head()
```

```python
points = []
for sheet in sheets:
    points += list(df_data[sheet].iloc[:, 0])
    
points
```

```python
len(points)
```

```python
code_map = {
    "LV": "Liv",
    "BI": "BL",
    "Bl": "BL",    
    "St": "ST",
    "DU": "Du",
}
```

```python
for _, point in enumerate(points):
    for code in code_map:
        if code in point:
            points[_] = point.replace(code, code_map[code])
            
points
```
